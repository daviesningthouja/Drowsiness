```python
import os
import tensorflow as tf
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import numpy as np
import scipy as spicy

# Define paths
checkpoint_dir = "D:/Davies Ningthoujam/My project/Ml Project/Driver drowsiness Detection/models"  
train_data_dir = 'D:/Davies Ningthoujam/My project/Ml Project/Driver drowsiness Detection/Data/Train'  
val_data_dir = 'D:/Davies Ningthoujam/My project/Ml Project/Driver drowsiness Detection/Data/Train'  

# Load the best model
best_model_path = os.path.join(checkpoint_dir, 'best_model.h5')  # Adjust 'best_model.h5' to your actual filename
best_model = load_model("D:\Davies Ningthoujam\My project\Ml Project\Driver drowsiness Detection\models")

# Compile the model (good practice if you need further fine-tuning)
best_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Set up data generators for evaluation
val_datagen = ImageDataGenerator(rescale=1./255)
val_data = val_datagen.flow_from_directory(
    val_data_dir,
    target_size=(80, 80),  
    batch_size=8,
    class_mode='categorical',
    shuffle=False  # Important for evaluation
)

# Get the true labels
y_true = val_data.classes  # True labels from the validation data

# Predict on validation data
y_pred = best_model.predict(val_data)
y_pred_classes = np.argmax(y_pred, axis=1)  # Convert predictions to class labels

# Calculate confusion matrix
cm = confusion_matrix(y_true, y_pred_classes)

# Display the confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=val_data.class_indices.keys())
disp.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix for Driver Drowsiness Detection")
plt.show()

# Evaluate the model on the validation dataset (optional)
loss_val, acc_val = best_model.evaluate(val_data)
print(f"Validation Accuracy: {acc_val}")
print(f"Validation Loss: {loss_val}")
```

```python
from sklearn.metrics import f1_score
# Calculate F1 score
f1 = f1_score(y_true, y_pred_classes, average='weighted')  # 'weighted' accounts for label imbalance

print(f"F1 Score: {f1}")
```

```python
import numpy as np
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay , precision_score, recall_score , f1_score
import matplotlib.pyplot as plt

test_datagen = ImageDataGenerator(rescale=1./255)

test_data = test_datagen.flow_from_directory(r'D:\Davies Ningthoujam\My project\Ml Project\Driver drowsiness Detection\eye',
                                target_size=(80,80),batch_size=8,class_mode='categorical', shuffle=False)

# Make predictions on the test data
y_pred = best_model.predict(test_data)
y_pred_classes = np.argmax(y_pred, axis=1)  # Convert probabilities to binary class labels

# Get the true labels
y_true = test_data.classes  # True labels from the test data

# Calculate the confusion matrix
cm = confusion_matrix(y_true, y_pred_classes)

# Calculate precision, recall, and F1 score
precision = precision_score(y_true, y_pred_classes, average='weighted')  # 'weighted' for multi-class
recall = recall_score(y_true, y_pred_classes, average='weighted')
f1 = f1_score(y_true, y_pred_classes)  # No need for 'average' in binary

# Print evaluation metrics
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f1}")

# Display the confusion matrix with custom class labels
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Close (Drowsy)', 'Open (Awake)'])
disp.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix for Driver Drowsiness Detection (Test Data)")
plt.show()


```
